```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
#columns cabin,ticket and number contains mixed data
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
#extract numerical data
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') #captures numerical part
df['cabin_categ'] = df['Cabin'].str[0] #captures the first letter

df.head()
```

```python
df['cabin_categ'].value_counts()
```

```python
df['cabin_categ'].value_counts().plot(kind='pie')
```

```python
#extract late bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                 errors='coerce',
                                 downcast='integer')

#extract first part of ticket as category
df['ticket_categ'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_categ'] = np.where(df['ticket_categ'].str.isdigit(),np.nan,
                            df['ticket_categ'])

df.head(20)
```
